```python
import pandas as pd
import numpy as np
from langdetect import detect
```

```python
df = pd.read_csv("data/morbid_metal_reviews.csv")
```

```python
df = df.drop_duplicates(
    subset=["recommendation_id"]
)

df = df.dropna(
    subset=["review_text"]
)
```

```python
if "language" not in df.columns:

    df["language"] = df["review_text"].apply(
        lambda x: detect(str(x))
    )
```

```python
df = df[df["language"] == "en"].copy()
```

```python
for topic in TOPICS:
    df[topic] = 0
```

```python
for idx, row in df.iterrows():

    review = str(
        row["review_text"]
    ).lower()

    for topic, keywords in TOPICS.items():

        if any(
            keyword in review
            for keyword in keywords
        ):
            df.at[idx, topic] = 1
```

```python
df.to_csv(
    "data/morbid_metal_processed.csv",
    index=False
)
```

```python
summary = []

for topic in TOPICS:

    total_mentions = df[topic].sum()

    positive_mentions = positive[topic].sum()

    negative_mentions = negative[topic].sum()

    summary.append({
        "Topic": topic,

        "Total_Mentions": total_mentions,

        "Mention_Rate_%": round(
            total_mentions / len(df) * 100,
            2
        ),

        "Positive_Mentions": positive_mentions,

        "Positive_%_Of_Total": round(
            positive_mentions / len(df) * 100,
            2
        ),

        "Negative_Mentions": negative_mentions,

        "Negative_%_Of_Total": round(
            negative_mentions / len(df) * 100,
            2
        ),

        "Positive_Rate_%": round(
            positive_mentions / total_mentions * 100,
            2
        ) if total_mentions > 0 else 0,

        "Negative_Rate_%": round(
            negative_mentions / total_mentions * 100,
            2
        ) if total_mentions > 0 else 0
    })

summary_df = pd.DataFrame(summary)

summary_df = summary_df.sort_values(
    by="Total_Mentions",
    ascending=False
)

summary_df
```
